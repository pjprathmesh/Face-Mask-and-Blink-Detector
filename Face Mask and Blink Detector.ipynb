# Face Mask

```python
import cv2
import numpy as np
import dlib

cap=cv2.VideoCapture(0)
detect=dlib.get_frontal_face_detector()
pred=dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
while 1:
    l1=[]
    _,frame=cap.read()
    
    faces=detect(frame)
    
    for face in faces:
       
        land=pred(frame,face)
#         print(land)
        for i in range(0,68):
            
            x=land.part(i).x
            y=land.part(i).y
            print(x)
            l1.append((x,y))
            
        arr=np.array(l1)
        hul=cv2.convexHull(arr)
        rec=cv2.boundingRect(hul)
        #(xx,yy,w,h)=rec  
        #cv2.rectangle(frame,(xx,yy),(xx+w,yy+h),(0,255,0),3)
        
        subdiv=cv2.Subdiv2D(rec)
        subdiv.insert(l1)
        tri_list=subdiv.getTriangleList()
        for i in tri_list:
            pt1=(i[0],i[1])
            pt2=(i[2],i[3])
            pt3=(i[4],i[5])
            cv2.line(frame,pt1,pt2,(0,255,0),1)
            cv2.line(frame,pt2,pt3,(0,255,0),1)
            cv2.line(frame,pt1,pt3,(0,255,0),1)
        
    cv2.imshow("frame",frame)
    key=cv2.waitKey(1)
    if key==27:
        cv2.destroyAllWindows()
        break 
```

# Blink Detector

```python
import cv2
import numpy as np
import dlib

cap=cv2.VideoCapture(0)
detect=dlib.get_frontal_face_detector()
pred=dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

while 1:
    l1=[]
    l2=[]
    _,frame=cap.read()
    faces=detect(frame)
    
    
    for face in faces:

        x1=face.left()
        y1=face.top()
        x2=face.right() 
        y2=face.bottom()
        l2=[[[x1,y1]],[[x1,y2]],[[x2,y2]],[[x2,y1]]]
        
        
        land=pred(frame,face)
        for i in [37,38,40,41]:
            
            x=land.part(i).x
            y=land.part(i).y
            l1.append([[x,y]])
    l2=np.array(l2)
    l1=np.array(l1)
    
    if len(l1)>0 and len(l2)>0:
        #cv2.drawContours(frame, [l1], 0,(120,147,34), 3)
        #cv2.drawContours(frame, [l2], 0,(120,147,34), 3)
        area_eye= cv2.contourArea(l1)
        area_face=cv2.contourArea(l2)
        face_to_eye_ratio=area_eye/area_face

        
        
        font = cv2.FONT_HERSHEY_SIMPLEX 
        org = (0, 50) 
        fontScale = 1 
        color =(0,255,0)
        thickness = 2
        text1 = cv2.putText(frame, "eye to face ratio: "+str(face_to_eye_ratio), org, font,  
                           fontScale, color, thickness, cv2.LINE_AA)  
        
        if face_to_eye_ratio<=0.0025:
            text2 = cv2.putText(frame,"Don't blink ", (50,450), font,  
                           fontScale, color, thickness, cv2.LINE_AA)  
    cv2.imshow("frame",frame)
    key=cv2.waitKey(1)
    if key==27:
        cap.release()
        break 
        
        
cv2.destroyAllWindows()
```
